```python
import copernicusmarine
import xarray as xr
import numpy as np
```

```python
DATASET_ID = "cmems_obs_si_arc_phy_my_L4-DMIOI_P1D-m" #id for Arctic Ocean - Sea and Ice Surface Temperature L4 REPROCESSED
VARIABLE_ICE = "sea_ice_fraction" # Sea Ice Concentration
VARIABLE_TEMP = "analysed_st" # SST

MIN_LAT = 58.0
MAX_LAT = 90.0
MIN_LON = -180.0
MAX_LON = 180.0

TARGET_LAT_RES = 2.0
TARGET_LON_RES = 2.0

start_years = np.arange(1982, 2033, 10)
end_years = np.arange(1991, 2032, 10)
```

```python
for i in range(len(start_years)-1):
    
    start_year = str(start_years[i])
    end_year = str(end_years[i])

    START_DATE = start_year + "-01-01"
    END_DATE = end_year + "-12-31"

    OUTPUT_FILENAME = start_year + "_" + end_year + "_" + "monthly_arctic_summary.nc"

    ds = copernicusmarine.open_dataset(
            dataset_id=DATASET_ID,
            variables=[VARIABLE_ICE, VARIABLE_TEMP],
            minimum_latitude=MIN_LAT,
            maximum_latitude=MAX_LAT,
            minimum_longitude=MIN_LON,
            maximum_longitude=MAX_LON,
            start_datetime=START_DATE,
            end_datetime=END_DATE,
            service="timeseries",
            username='edominguez',
            password='xpWnwb3k3@F7R!K'
    )
        
    ds_monthly = ds[[VARIABLE_ICE, VARIABLE_TEMP]].resample(time="ME").mean()
        
    lat_bins = np.arange(MIN_LAT, MAX_LAT + TARGET_LAT_RES, TARGET_LAT_RES)
    lon_bins = np.arange(MIN_LON, MAX_LON + TARGET_LON_RES, TARGET_LON_RES)

    ds_aggregated = ds_monthly.groupby_bins('latitude', bins=lat_bins, include_lowest=True).mean()
    ds_aggregated = ds_aggregated.groupby_bins('longitude', bins=lon_bins, include_lowest=True).mean()
        
    ds_aggregated['latitude_bins'] = ds_aggregated['latitude_bins'].to_index().mid
    ds_aggregated['longitude_bins'] = ds_aggregated['longitude_bins'].to_index().mid
        
    ds_aggregated = ds_aggregated.rename({
        'latitude_bins': 'latitude_1deg',
        'longitude_bins': 'longitude_1deg'
    })
    print(f"Successfully aggregated data for {start_year}–{end_year}")
    
    ds_aggregated.compute().to_netcdf(OUTPUT_FILENAME)

    print(f"Successfully exported file to {OUTPUT_FILENAME}")
```
